```python
import urllib.request
from bs4 import BeautifulSoup
foxResponse = urllib.request.urlopen("http://www.foxnews.com")
fox = foxResponse.read()
foxSoup = BeautifulSoup(fox, "lxml")
```

```python
for title in foxSoup.find_all("h2", class_="title"):
    print(title.text)
```

```python
import urllib.request
from bs4 import BeautifulSoup
cbcResponse = urllib.request.urlopen("http://www.cbc.ca")
cbc = cbcResponse.read()
cbcSoup = BeautifulSoup(cbc, "lxml")
```

```python
for headline in cbcSoup.find_all("h3", class_="headline"):
    print(headline.text)
```
